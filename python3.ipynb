3.Palindrome Check & Digit Occurence Count:Develop a python program to check whether a given number is palindrome or not and also count the number of occurences of each digit in the input number.


```python
number = input("Enter a number: ")
if number == number[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
for i in range(10):
    count = number.count(str(i))
    if count > 0:
        print(f"{i} appears {count} times")
```
